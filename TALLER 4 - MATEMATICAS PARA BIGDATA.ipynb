```python
# En R**2
a1, a2, a3 = sympy.symbols('a1, a2, a3')
from sympy import *
A = sympy.Matrix(((12, -3, 4), (5, 0, -8)))
A
```

```python
print(sympy.solve_linear_system(A, a1, a2, a3), 'El conjunto de vectores si genera espacio Vectorial en R2')
```
